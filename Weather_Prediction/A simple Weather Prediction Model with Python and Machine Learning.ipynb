### Predicting  the next day's tempereature(tmax

```python
#Creating a target column with  shifted temperature
weather_ffill['target']=weather_ffill.shift(-1)['tmax']
```

```python
#Applying the model
from sklearn.linear_model import Ridge
```

```python
#initializing the model

rr=Ridge(alpha=.1)
```

```python
#Create a list of predictor columns
predictors=weather_ffill.columns[~weather_ffill.columns.isin(['target','name',"station",'latitude', 'longitude', 'elevation',
                                                             'prcp_attributes','tmax_attributes','tavg_attributes',
                                                              'tmin_attributes'  ])]
```

```python
predictors
```

```python
#Genereate predictions for the entire set of data expect the first 10 years
def backtest(weather, model, predictors, start=3650, step=90):
    all_predictions=[]

    for i in range(start, weather.shape[0], step):
        train=weather.iloc[:i,:]
        test=weather.iloc[i:(i+step), :]
        
        model.fit(train[predictors], train['target'])
        
        preds=model.predict(test[predictors])
        
        preds=pd.Series(preds, index=test.index)
        combined=pd.concat([test['target'],preds], axis=1)
        
        combined.columns=['actual', 'prediction']
        
        combined['diff']=(combined['prediction']- combined['actual']).abs()
        
        all_predictions.append(combined)
    
    return pd.concat(all_predictions)
```

```python
predictions=backtest(weather_ffill, rr, predictors)
```

```python
predictions
```

```python
#error metrics
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(predictions['actual'], predictions['prediction'])
```

```python
#Generating predictors to improve our accuracy
```

```python
def pct_diff(old, new):
    return (new - old)/old

def compute_rolling(weather, horizon, col):
    label=f"rolling_{horizon}_{col}"
    
    weather[label]=weather[col].rolling(horizon).mean()
    weather[f"{label}_pct"]=pct_diff(weather[label], weather[col])
    return weather

rolling_horizons=[3,14]

for horizon in rolling_horizons:
    for col in ['tmax', 'tmin', 'prcp']:
        weather_ffill=compute_rolling(weather_ffill, horizon, col)
```

```python
weather_ffill
```

```python
#remove the first 14 rows  because of missing data
weather_ffill=weather_ffill.iloc[14:,:]
```

```python
weather_ffill
```

```python
#because of the percentage column we will fill all the missing values with zero
weather_ffill=weather_ffill.fillna(0)
```

```python
#Adding more predictors
def expand_mean(df):
    return df.expanding(1).mean()

for col in ['tmax', 'tmin', 'prcp']:
    weather_ffill[f"month_avg{col}"]=weather_ffill[col].groupby(weather_ffill.index.month, group_keys=False).apply(expand_mean)
    weather_ffill[f"day_avg{col}"]=weather_ffill[col].groupby(weather_ffill.index.day_of_year, group_keys=False).apply(expand_mean)
```
